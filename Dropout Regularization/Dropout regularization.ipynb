```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("Connectionist Bench.csv", header=None)
df.sample(5)
```

```python
df.columns
```

```python
X=df.drop(60, axis='columns').values
y=df[60]

```

```python
y=pd.get_dummies(y, drop_first=True) #R-->1, M--->0
y.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=1)
X_train, y_train
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model= keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
    
])

model.compile(
                loss='binary_crossentropy',
                optimizer='adam',
                metrics=['accuracy']

)

model.fit(X_train,y_train, epochs=100, batch_size=8) #batch_size= mini batch gradient desent

```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred= np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_pred))
```

```python
#Dropout layer

modeld= keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.6),
    keras.layers.Dense(1, activation='sigmoid')
    
])

modeld.compile(
                loss='binary_crossentropy',
                optimizer='adam',
                metrics=['accuracy']

)

modeld.fit(X_train,y_train, epochs=200, batch_size=8) #batch_size= mini batch gradient desent
```

```python
modeld.evaluate(X_test,y_test)
```
